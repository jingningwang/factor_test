```python
# 已经进行测试
# 传入一个df，包含trade_price, tarde_volume 两列即可
def cal_wei_price(df):
    total_sum = 0
    for i in range(df.shape[0]):
        temp_sum = df.iloc[i]['trade_price'] * df.iloc[i]['trade_volume']
        total_sum += temp_sum 
    sum_vol = sum(df[df['trade_price']!=0]['trade_volume'])
    if sum_vol == 0:
        wei_price = 0
    else:
        wei_price = total_sum / sum_vol
    return wei_price
```

```python
def avg(df_3):
    if df_3['time_win']==None:
            return None
        
    # 如果中心是0
    if df_3['time_win'][10][0] == 0:
        for j in range(-10,21):
            df_3['time_win'][j+10][j] +=100
        return df_3['time_win']
            
    # 中心不是0
    else:
        for j in range(-10,0):
                df_3['time_win'][j+10][j] = df_3['time_win'][j+10][j] / df_3['time_win'][10][0]
                df_3['time_win'][j+10][j] = 100 * df_3['time_win'][j+10][j]
                
        for j in range(1,21):
                df_3['time_win'][j+10][j] = df_3['time_win'][j+10][j] / df_3['time_win'][10][0]
                df_3['time_win'][j+10][j] = 100 * df_3['time_win'][j+10][j]
        df_3['time_win'][10][0] = 100
        return df_3['time_win']
```

```python
def cal_p_mat_type_8(order):
    mat = []
    for i in range(order.shape[0]):
        if order.loc[i]['type_agg'] == 'type8' and order.loc[i]['time_win'] is not None:
            # mat是一个每个元素都是字典的列表
            # 字典的 key 是 
            # print(order.loc[i]['time_win_avg'])
            mat.append(order.loc[i]['time_win_avg'])
            
    # 对 p 矩阵进行处理,s实现了转置
    mat_p = [[0] * len(mat) for _ in range(31)]

    # j是类别数，k是样本数
    for j in range(31):
        for k in range(len(mat)):
            mat_p[j][k] = mat[k][j][j-10]
    return mat_p
```
