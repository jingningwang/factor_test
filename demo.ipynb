```python
def select_type_1(df):
    # 将快照表的买价卖价筛出来
    for k in range(1,6):
        df['0bid'+str(k)] = df['buy_delegations'].apply(lambda x: x[k-1]['price'])
        df['1bid'+str(k)] = df['0bid'+str(k)].shift(-1)
        df['0ask'+str(k)]=df['sell_delegations'].apply(lambda x: x[k-1]['price'])
        df['1ask'+str(k)] = df['0ask'+str(k)].shift(-1)

    df['da'] = df['1ask1'] - df['0ask1']
    df['db'] = df['0bid1'] - df['1bid1']
    # 条件判断
    condition_b_type_b_5 = (( df['1ask1'] > df['0ask5']) & (df['da']>100)) | ((
                df['1bid1'] < df['0bid5']) & (df['db'] > 100))
    condition_b_type_b_4 = (( df['1ask1'] > df['0ask4']) & (df['da']>100)) | ((
                df['1bid1'] < df['0bid4']) & (df['db'] > 100))
    condition_b_type_b_3 = (( df['1ask1'] > df['0ask3']) & (df['da']>100)) | ((
                df['1bid1'] < df['0bid3']) & (df['db'] > 100))
    condition_b_type_b_2 = (( df['1ask1'] > df['0ask2']) & (df['da']>100)) | ((
                df['1bid1'] < df['0bid2']) & (df['db'] > 100))
    condition_b_type_a = ((df['1ask1'] > df['0ask1']) & (df['da']>100)) | ((
                            df['1bid1'] < df['0bid1']) & (df['db'] > 100))


    condition_b_type_one = ((df['1ask1'] > df['0ask1']) & (df['da']==100)) | ((
                            df['1bid1'] < df['0bid1']) & (df['db'] == 100))
    condition_b_type_zero = ((df['1ask1'] == df['0ask1']) & (df['da']==0)) | ((
                            df['1bid1'] == df['0bid1']) & (df['db'] == 0))

    df['type_agg'] = np.select(
        [condition_b_type_b_5, condition_b_type_b_4, condition_b_type_b_3, condition_b_type_b_2, condition_b_type_a, condition_b_type_one,condition_b_type_zero],
        ['type_b5', 'type_b4', 'type_b3', 'type_b2','type_a','type_one','type_zero'],default='other' )
    return df
```

```python
def select_type_2(df):
    df.set_index('tick_time',inplace=True)
    df = df.resample('T').last()
    df = df[(df.index.time <= pd.to_datetime('11:30').time()) | (df.index.time >= pd.to_datetime('13:00').time())]
    df.fillna(method='ffill', inplace = True)
    df['bid1'] = df['buy_delegations'].apply(lambda x: x[0]['price'])
    df['ask1'] = df['sell_delegations'].apply(lambda x: x[0]['price'])
    df['price'] = (df['bid1']+df['ask1'])/2
    df['price_before'] = df['price'].shift(1)
    # 计算一个时间点的对数收益率
    df['log_return'] = np.log(df['price'] / df['price_before'])
    # 计算累计率
    for k in range(1,61):
        df['logret_before_' + str(k) + 'minutes'] = df['log_return'].shift(k)
    df['cum_return'] = df[['logret_before_' + str(k) + 'minutes' for k in range(1,61)]].sum(axis=1)
    # 计算平均波动率
    avg_col = ['logret_before_' + str(k) + 'minutes' for k in range(1,16)]
    df['avg_cumret'] = (df[avg_col]**2).sum(axis=1)
    df['avg_cumret'] = np.sqrt(df['avg_cumret'])
    return df
```
